```python
# Подключаем пакеты:
import sympy as sp
sp.init_printing()
```

```python
x = sp.symbols('x') # Переменная x теперь содержит символ 'x'
```

```python
f = (x**3+4)/x**2 # Выражение для функции
f
```

```python
# Инициализация отображения графиков в блокноте:
%matplotlib inline
sp.plot(f, (x, -1, 1)); #sympy содержит свою функцию для быстрого построения графиков
```

```python
# Вычислить значение при x = 0:
f.subs(x, 0)
# График функции не пересекает ось OY
```

```python
sp.solve(f, x) #Корни уравнения f(x) = 0
```

 # Производная


```python
df = f.diff(x, 1)  #Производная
df
```

```python
sp.plot(df, (x, -1, 3)); #sympy содержит свою функцию для быстрого построения графиков
```

```python
# Вычислить значение производной при x = 0:
df.subs(x, 0)
```

```python
sp.solve(df, x) #Корни уравнения производной f(x) = 0
```

```python
df_2 = df.diff(x, 1)  # Вторая производная
df_2
```

```python
sp.limit(df,x,0)
```
